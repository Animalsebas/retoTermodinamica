```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
sizex = 32
sizey = 32
```

```python
def draw(matrix):
  rows, cols = matrix.shape
  fig, ax = plt.subplots()

  for i in range(rows):
      for j in range(cols):
          color = 'red' if matrix[i, j] == 1 else 'blue'
          circle = plt.Circle((j, rows - i - 1), 0.4, color=color, fill=True)
          ax.add_patch(circle)

  ax.set_xlim(-0.5, cols - 0.5)
  ax.set_ylim(-0.5, rows - 0.5)
  ax.set_xticks(np.arange(cols))
  ax.set_yticks(np.arange(rows))
  ax.set_xticklabels(range(cols))
  ax.set_yticklabels(range(rows))

  ax.grid(which='both', linestyle='--', color='gray', linewidth=0.5)

  plt.title('Gráfica de estado')
  plt.show()
```

```python
configuracion_inicial = np.random.choice([-1,1], size=(sizex, sizey))
```

```python
draw(configuracion_inicial)
```

```python
def graphEnergyMag(temperatura,promEnergy,promMag):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

    # Plot the first subplot
    axes[0].plot(promEnergy, label='<E>', color='r')
    axes[0].set_title(("Energía con T = "+str(temperatura)))
    axes[0].set_xlabel("Secuencia")
    axes[0].set_ylabel('<E>')
    axes[0].legend()

    # Plot the second subplot
    axes[1].plot(promMag, label='<|M|>', color='b')
    axes[1].set_title(("Magnetización con T = "+str(temperatura)))
    axes[1].set_xlabel("Secuencia")
    axes[1].set_ylabel('<|M|>')
    axes[1].legend()
```

```python
temperatura = 1.8
with open("32x32 FINAL/secuencia_estados1-8", "rb") as fp:   # Unpickling
    secuencias18 = pickle.load(fp)
with open("32x32 FINAL/magnetizaciones1-8", "rb") as fp:   # Unpickling
    promMagList18 = pickle.load(fp)
with open("32x32 FINAL/energias1-8", "rb") as fp:   # Unpickling
    promEnergyList18 = pickle.load(fp)
graphEnergyMag(temperatura,promEnergyList18,promMagList18)
print("Promedio de la energía: "+str(np.mean(promEnergyList18)))
print("Promedio de la magnetización: "+str(np.mean(promMagList18)))
```

```python
randomSelection = secuencias18[-1][-1]
draw(randomSelection)
total_elements = sizex*sizey
num_iterations = 10
percentage_minus_1 = []
percentage_1 = []

current_selection = secuencias18[2][-1]
count_minus_1 = np.count_nonzero(current_selection == -1)
count_1 = np.count_nonzero(current_selection == 1)

percentage_minus_1.append((count_minus_1 / total_elements) * 100)
percentage_1.append((count_1 / total_elements) * 100)

print(f'Porcentaje de spin -1: {np.mean(percentage_minus_1):.2f}%')
print(f'Porcentaje de spin 1: {np.mean(percentage_1):.2f}%')
```

```python
def energiaEstado(matrix):
  
  def energy(matrix,x,y):
    def get_neighbors(matrix, x, y):
      rows, cols = len(matrix), len(matrix[0])
      # Vecinos con fronteras periodicas (toroidales)
      vecinos = [
          #((x - 1) % rows, y),  # Arriba
          ((x + 1) % rows, y),  # Abajo
          #(x, (y - 1) % cols),  # Izquierda
          (x, (y + 1) % cols)   # Derecha
      ]

      
      spinsVecinos = [matrix[i][j] for i, j in vecinos]

      spinsVecinos.insert(0, matrix[x][y])

      return spinsVecinos
    spinsVecinos = get_neighbors(matrix, x, y)
    E = 0
    for i in range(2):
      E+=spinsVecinos[0]*spinsVecinos[i+1]
    return E
  N, M = matrix.shape
  E = 0
  for i in range(N):
    for j in range(M):
      E += energy(matrix,i,j)
  N = 32*32
  E = (-1/N)*E
  return E

configuracion_inicial = np.ones((32, 32), dtype=int)
print(energiaEstado(configuracion_inicial))
```
